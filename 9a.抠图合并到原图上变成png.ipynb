```python
import cv2
import numpy as np
import os
```

```python
path = '/Users/roger/deeplearning/bigDatasets/faces/dataset/face/raw/01train'
path_alpha = '/Users/roger/deeplearning/bigDatasets/faces/dataset/face/raw/01train_alpha'
path_png = '/Users/roger/deeplearning/bigDatasets/faces/dataset/face/raw/01png'
for filename in os.listdir(path):
    print("do",filename)
    fullpath = os.path.join(path,filename)
    fullpath_alpha = os.path.join(path_alpha,filename[:-4]+'.bmp')
    fullpath_png = os.path.join(path_png,filename[:-4]+'.png')
    img=cv2.imread(fullpath)
    img_alpha=cv2.imread(fullpath_alpha)
    img_alpha = img_alpha/2+128 #不透明的地方如果alpha=0，png回自动优化这些点，使得rbg=0
    img_alpha = img_alpha[:,:,0].reshape((img_alpha.shape[0], img_alpha.shape[1], 1)) #取alpha = r
    png = np.concatenate((img,img_alpha),axis=2)
    cv2.imwrite(fullpath_png, png)  
    
    #TODO: 如果直接进skipnet,w,h 如果是奇数，需要调整为偶数


```
